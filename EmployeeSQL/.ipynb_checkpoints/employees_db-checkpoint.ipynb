```python
import pandas as pd
import matplotlib.pyplot as plt
from config import username, password

plt.style.use('fivethirtyeight')
```

```python
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/Employee_sql')
connection = engine.connect()
```

```python
# Import the employees table
employees = pd.read_sql('select * from employees', connection, parse_dates=['birth_date', 'hire_date'])
```

```python
employees.head()
```

```python
# Import the Departments table
departments = pd.read_sql('select * from departments', connection)
departments.head()
```

```python
# Import the salaries table
salaries = pd.read_sql('select * from salaries', connection)
salaries.head()
```

```python
# Import the department manager table
dept_manager = pd.read_sql('select * from dept_manager', connection, parse_dates=['from_date', 'to_date'])
dept_manager.head()
```

```python
# Import the titles table
titles = pd.read_sql('select * from titles', connection, parse_dates=['from_date', 'to_date'])
titles.head()
```

```python
# Import the departments employees table
dept_emp = pd.read_sql('select * from dept_emp', connection, parse_dates=['from_date', 'to_date'])
dept_emp.tail()
```

```python
# Join employees and salaries

employee_salaries = employees.merge(salaries, on='emp_no')
employee_salaries.head()
```

```python
employee_title_salaries = employee_salaries.merge(titles, on='emp_no')
employee_title_salaries.head()
```

```python
salaries_df = employee_title_salaries[['title','salaries']]
salaries_df.head()
```

```python
titles_group = salaries_df.groupby('title')['salaries'].mean().round(2).sort_values(ascending=False)
titles_group
```

```python
titles_group.plot(kind='bar')
plt.xlabel("Salary Ranges")
plt.ylabel("Frequency")
```
